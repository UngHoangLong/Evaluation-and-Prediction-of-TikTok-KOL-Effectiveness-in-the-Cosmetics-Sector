# Note 
- Bởi vì t test hoạt động dựa trên giả định rằng phân phối của hai nhóm là gần chuẩn và phương sai không quá khác biệt. Vậy nên khi nhận thấy rằng phân phối của 2 nhóm này không tuân theo phân phối chuẩn, việc sử dụng t-test để thực hiện kiểm định giả thuyết là không phù hợp. 
    + Đưa ra một giải pháp mới đó là sử dụng một phương pháp kiểm định mới - một kiểm định phi tham số (non - parameter test ) kiểm định Mann-Whitney U (còn gọi là Wilcoxon rank-sum test), vì nó không yêu cầu giả định về phân phối chuẩn hoặc phương sai bằng nhau.
    + Mann-Whitney U so sánh phân phối (thường là trung vị) của hai nhóm

- Tham sô (parametric) là gì ? 
    + Phương pháp tham số là các phương pháp thống kê giả định rằng dữ liệu tuân theo một phân phối xác định, thường là phân phối chuẩn (normal distribution), và sử dụng các tham số của phân phối này (như trung bình, độ lệch chuẩn) để mô tả dữ liệu.
    + Đặc điểm:
        Giả định phân phối: Dữ liệu được giả định tuân theo một phân phối cụ thể (thường là chuẩn).
        Tham số: Sử dụng các tham số như trung bình ($\mu$) và phương sai ($\sigma^2$) để mô hình hóa.
        Ví dụ kiểm định:

            t-test (so sánh trung bình).
            ANOVA (phân tích phương sai).
            Hồi quy tuyến tính (linear regression).
    + Ưu điểm:
        Hiệu quả cao (powerful) khi giả định được thỏa mãn, vì tận dụng thông tin về phân phối.
        Kết quả chính xác hơn với mẫu lớn và dữ liệu chuẩn.
    + Nhược điểm:
        Nhạy cảm với vi phạm giả định (như dữ liệu không chuẩn hoặc phương sai không bằng nhau).
        Không phù hợp với dữ liệu ordinal hoặc dữ liệu có phân phối bất thường.
- Phi tham số (Non-parametric) là gì?
    + Phương pháp phi tham số là các phương pháp thống kê không giả định về hình dạng cụ thể của phân phối dữ liệu (như phân phối chuẩn). Thay vào đó, chúng dựa trên thứ hạng (ranks), trung vị, hoặc các đặc điểm khác của dữ liệu.

    + Đặc điểm:

        Không giả định phân phối: Không yêu cầu dữ liệu phải tuân theo phân phối chuẩn hoặc bất kỳ phân phối nào khác.
        Dựa trên thứ hạng hoặc cấu trúc dữ liệu: Thường sử dụng thứ hạng (rank) hoặc dấu hiệu (sign) thay vì giá trị gốc.
        Ví dụ kiểm định:
            Mann-Whitney U (so sánh hai nhóm độc lập).
            Wilcoxon signed-rank test (so sánh hai nhóm liên quan).
            Kruskal-Wallis test (so sánh nhiều nhóm).
        Ưu điểm:
            Linh hoạt, phù hợp với dữ liệu không chuẩn, dữ liệu ordinal, hoặc dữ liệu có outlier.
            Ít nhạy cảm với vi phạm giả định về phân phối.
        Nhược điểm:
            Ít nhạy (less powerful) hơn phương pháp tham số khi dữ liệu thỏa mãn giả định chuẩn.
            Có thể yêu cầu kích thước mẫu lớn hơn để đạt độ chính xác tương đương.

# I. Nhóm kiểm định t-test (hai mẫu độc lập)

```python
from scipy.stats import shapiro
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
from scipy.stats import levene

def check_normal_distribution(group1, group2):
    #print("Kiểm tra phân phối chuẩn:")
    #print("group_2:", shapiro(group1))
    #print("group_1:", shapiro(group2))
    if (shapiro(group1).pvalue < 0.05) or (shapiro(group2).pvalue < 0.05):
        print('Dữ liệu có phân phối không chuẩn')
        return 0
    else:
        print('Dữ liệu có phân phối chuẩn')
        return 1
    

def t_test(group1, group2):
    t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
    print(f'T-statistic: {t_stat:.4f}')
    print(f'P-value: {p_value:.4f}')
    if p_value < 0.05:
        print("Bác bỏ H₀")
    else:
        print("Không đủ bằng chứng bác bỏ H₀")


def check_mannwhitneyu(group1, group2):
    stat, p_value = mannwhitneyu(group1, group2)
    #print(f'Mann-Whitney U statistic: {stat:.4f}')
    print(f'P-value: {p_value:.4f}')
    if p_value < 0.05:
        print("Bác bỏ H₀")

    else:
        print("Không đủ bằng chứng bác bỏ H₀")
    


def variance(group1, group2):
    var = levene(group1, group2)
    if var.pvalue < 0.05:
        print('Phương sai không đồng đều')
        return 0
    else:
        print('Phương sai đồng đều')
        return 1
def visualization(col):
    sns.histplot(data = data, x = col, bins = 30)
    plt.show()
def calculate_effect_size(group1, group2):
    """
    Tính kích thước hiệu ứng (r) dựa trên kiểm định Mann-Whitney U.
    
    Parameters:
    - group1: Danh sách hoặc mảng dữ liệu của nhóm 1 (group_true)
    - group2: Danh sách hoặc mảng dữ liệu của nhóm 2 (group_false)
    
    Returns:
    - r: Kích thước hiệu ứng
    - effect_interpretation: Giải thích mức độ hiệu ứng
    """
    # Thực hiện kiểm định Mann-Whitney U
    stat, p_value = mannwhitneyu(group1, group2, alternative='two-sided')
    
    # Tính Z-score từ thống kê U
    N1 = len(group1)
    N2 = len(group2)
    N = N1 + N2
    U_expected = (N1 * N2) / 2
    U_diff = abs(stat - U_expected)
    z_score = U_diff / np.sqrt((N1 * N2 * (N + 1)) / 12)
    
    # Tính kích thước hiệu ứng r
    r = z_score / np.sqrt(N)
    
    # Giải thích mức độ hiệu ứng
    if abs(r) < 0.1:
        effect_interpretation = "Hiệu ứng rất nhỏ"
    elif 0.1 <= abs(r) < 0.3:
        effect_interpretation = "Hiệu ứng nhỏ"
    elif 0.3 <= abs(r) < 0.5:
        effect_interpretation = "Hiệu ứng trung bình"
    else:
        effect_interpretation = "Hiệu ứng lớn"
    return r, effect_interpretation
```

## Effect_iterpretation 
* Ý nghĩa của từng mức độ trong ngữ cảnh của kích thước hiệu ứng $ r $ từ kiểm định Mann-Whitney U 

1. Hiệu ứng nhỏ (r ~ 0.1)
- Ý nghĩa : Sự khác biệt hoặc ảnh hưởng giữa hai nhóm là rất nhẹ, gần như không đáng kể trong thực tế. Điều này có nghĩa là, mặc dù có thể có sự khác biệt thống kê (nếu p-value < 0.05), nhưng sự khác biệt đó không đủ lớn để có ý nghĩa thực tiễn.
2. Hiệu ứng trung bình (r ~ 0.3)
- Ý nghĩa: Sự khác biệt hoặc ảnh hưởng giữa hai nhóm có mức độ vừa phải, đủ để nhận thấy trong thực tế và có thể có ý nghĩa ứng dụng. Đây là mức độ cho thấy một mối quan hệ hoặc khác biệt rõ ràng hơn so với hiệu ứng nhỏ.
3. Hiệu ứng lớn (r ~ 0.5 hoặc cao hơn)
- Ý nghĩa: Sự khác biệt hoặc ảnh hưởng giữa hai nhóm rất rõ rệt và có ý nghĩa thực tiễn cao. Điều này cho thấy biến số độc lập (như "shareCount_pr") có tác động mạnh đến biến phụ thuộc (như "EngagementRateOnPRPost_target").

- Khi kết luận được hai mẫu có sự khác biệt về mặt thống kê (thông qua t-test hoặc Mann-Whitney U), nếu muốn đi sâu vào xem mức độ khác biệt đó như thế nào, việc tiếp theo là đánh giá mức độ khác biệt (effect size). Effect size giúp bạn hiểu rõ mức độ thực tiễn (practical significance) của sự khác biệt, thay vì chỉ dựa vào giá trị p (statistical significance).
1. Effect Size là gì?
- Effect size là một thước đo định lượng mức độ khác biệt giữa hai nhóm, không phụ thuộc vào kích thước mẫu. Nó giúp trả lời câu hỏi: "Sự khác biệt giữa hai nhóm lớn đến mức nào?" Một số thước đo effect size phổ biến:

    Cohen’s d: Dùng cho t-test (phương pháp tham số).
    
    r (rank-biserial correlation): Dùng cho Mann-Whitney U (phương pháp phi tham số).

```python
from scipy.stats import kruskal
from scikit_posthocs import posthoc_dunn
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def variance(group1, group2, group3):
    stat_levene, p_levene = levene(group1, group2, group3)
    if p_levene < 0.05:
        print("Phương sai không đồng đều giữa các nhóm.")
        return 0
    else:
        print("Phương sai đồng đều giữa các nhóm.")
        return 1
    
def check_normal_distribution(group1, group2, group3):
    if (shapiro(group1).pvalue < 0.05) or (shapiro(group2).pvalue < 0.05) or ((shapiro(group3).pvalue < 0.05)):
        print('Dữ liệu có phân phối không chuẩn')
        return 0
    else:
        print('Dữ liệu có phân phối chuẩn')
        return 1

def anova_test(group1, group2, group3):
    stat_anova, p_anova = f_oneway(group1, group2, group3)
    print("\nKết quả ANOVA một chiều:")
    print(f"F-statistic: {stat_anova:.4f}")
    print(f"P-value: {p_anova:.4f}")
    if p_anova < 0.05:
        print("Bác bỏ H₀: Có ít nhất một nhóm có trung bình tỷ lệ tương tác khác với các nhóm còn lại.")
        # Post-hoc test (Tukey’s HSD) nếu bác bỏ H₀
        print("\nKiểm định post-hoc (Tukey’s HSD):")
        data['group'] = pd.cut(data['total_posts_pr'], 
                               bins=[-float('inf'), 20, 50, float('inf')], 
                               labels=['Low', 'Medium', 'High'], 
                               right=False)
        tukey = pairwise_tukeyhsd(endog=data['EngagementRateOnPRPost_target'], 
                                  groups=data['group'], 
                                  alpha=0.05)
        print(tukey)
    else:
        print("Không đủ bằng chứng bác bỏ H₀: Trung bình tỷ lệ tương tác là như nhau giữa các nhóm.")
    

# kiểm định phi tham số: Kruskal-Wallis test
'''
def kruskal_test(group1, group2, group3, col):
    stat_kruskal, p_kruskal = kruskal(group1, group2, group3)
    print(f"Kruskal-Wallis statistic: {stat_kruskal:.4f}")
    print(f"P-value: {p_kruskal:.4f}")
    if p_kruskal < 0.05:
        print("Bác bỏ H₀oooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooo")
        # Kiểm định hậu nghiệm 
        print("\nKiểm định post-hoc (Dunn’s test):")
        dunn_result = posthoc_dunn(data, val_col=col, 
                                   group_col='group', p_adjust='bonferroni')
        print(dunn_result)
    else:
        print("Không đủ bằng chứng bác bỏ H₀.")
'''
def kruskal_test(group1, group2, group3, col):
    stat_kruskal, p_kruskal = kruskal(group1, group2, group3)
    #print(f"\nKruskal-Wallis statistic: {stat_kruskal:.4f}")
    print(f"P-value: {p_kruskal:.4f}")
    if p_kruskal < 0.05:
        print("Bác bỏ H₀: Có ít nhất một nhóm có phân phối tỷ lệ tương tác khác.")
        # Kiểm định hậu nghiệm (Dunn’s test)
        print("\nKiểm định post-hoc (Dunn’s test):")
        dunn_result = posthoc_dunn(data, val_col=col, group_col='group', p_adjust='bonferroni')
        print(dunn_result)
    else:
        print("Không đủ bằng chứng bác bỏ H₀: Phân phối tỷ lệ tương tác là như nhau giữa các nhóm.")
```
